```python
def calcula_desconto(valor, desconto):
    return valor - (valor * desconto / 100)

print(calcula_desconto(100, 10))
```

```python
def classifica_nota(nota):
    if nota >= 7:
        return "Aprovado"
    elif nota >= 5:
        return "Recuperação"
    else:
        return "Reprovado"

print(classifica_nota(6.5))
```

```python
numeros = [10, -3, 0, 5, -1, 0, 8, -7]

positivos = negativos = zeros = 0

for n in numeros:
    if n > 0:
        positivos += 1
    elif n < 0:
        negativos += 1
    else:
        zeros += 1

print(f"Positivos: {positivos}, Negativos: {negativos}, Zeros: {zeros}")
```
